```python
import pandas as pd
import glob
import os

path = '/home/tejaswik/code/sweetpatata/Bike-or-nei/raw_data/wetransfer_raw-data_2022-11-28_1515/data-bike/'
all_files = glob.glob(os.path.join(path , "*.csv"))

list = []

for filename in all_files:
    dframe = pd.read_csv(filename, index_col=None, header=0)
    dframe['start_date'] = [d.date() for d in pd.to_datetime(dframe['started_at'])]
    df1 = dframe[["started_at","start_date","end_station_id"]].groupby(["end_station_id",'start_date'],as_index = False).count()\
    .rename(columns={"end_station_id":"station_id","start_date":"date","started_at":"no_of_trips_in"})
    df2 = dframe[["started_at","start_date","start_station_id"]].groupby(["start_station_id",'start_date'],as_index = False).count()\
    .rename(columns={"start_station_id":"station_id","start_date":"date","started_at":"no_of_trips_out"})
    df4 = df1.merge(df2,on = ['station_id','date'])
    
    list.append(df4)


len(list)
```

```python
frame = pd.concat(list, ignore_index=True)
frame.shape
```
