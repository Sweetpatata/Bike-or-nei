```python
files = glob.glob('/Users/nhuphuong/code/sweetpatata/bike-or-nei/raw_data/*.csv')
files
```

```python
df = pd.DataFrame()
for file in files:
    df = pd.concat([df, pd.read_csv(file, parse_dates=['started_at', 'ended_at'])])
```

```python
len(df)
```

```python
df.head()
```

```python
df.info()
```

```python
df['start_date'] = df['started_at'].dt.date
df['end_date'] = df['ended_at'].dt.date
```

```python
df_out = df.groupby(['start_station_id', 'start_date']).agg('count')
```

```python
df_out = df_out[['started_at']]
```

```python
df_out.columns = ['Number_of_trips_out']
```

```python
df_in = df.groupby(['end_station_id', 'end_date']).agg('count')
```

```python
df_in = df_in[['started_at']]
```

```python
df_in.columns = ['Number_of_trips_in']
```

```python
df_all = df_in.merge(df_out, left_index=True, right_on=['start_station_id', 'start_date'])
```

```python
df_all.reset_index(inplace=True)
```

```python
df_all.columns = [['Station_Id', 'Date', 'Trips_in', 'Trips_out']]
```

```python
df_all.columns = [name[0] for name in df_all.columns]
```

```python
df_all['Out_Inn'] = df_all['Trips_out'] - df_all['Trips_in']
```

```python
len(df_all)
```

```python
df_all['Date'] = pd.to_datetime(df_all['Date'])
```

```python
df_all.info()
```
